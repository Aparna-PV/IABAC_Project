```python
plt.figure(figsize=(25,25))
p=1
for i in data.iloc[:,:27]:
    ax=plt.subplot(7,4,p)
    sns.histplot(x=data[i],kde=True,hue=data.PerformanceRating,palette = ['blue','orange'])
    p+=1
plt.tight_layout()
```

# BIVARIATE ANALYSIS

```python
plt.figure(figsize=(25, 25), facecolor="white")
p = 1
for i in data.iloc[:,:27]:
        ax = plt.subplot(7, 4, p)
        sns.countplot(x=data[i], hue=data["PerformanceRating"])
        plt.xlabel(i, fontsize=25)
        plt.ylabel("PerformanceRating", fontsize=25)
        p+=1
plt.tight_layout()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x="EducationBackground",y="PerformanceRating",data=data,hue="EducationBackground")
plt.title("Education VS Performance")
plt.tight_layout
```

### INSIGHT

Among individuals with other educational backgrounds outside of marketing, life science, human resources, medical, and technical fields,shows higher performance levels.

```python
plt.figure(figsize=(10,5))
sns.barplot(x="MaritalStatus",y="PerformanceRating",data=data,hue="MaritalStatus")
plt.title("Marital VS Performance")
plt.tight_layout
```

### INSIGHT

- People with different marital statuses regularly exhibit comparable performance levels. Regardless of their marital status—married, single, or in divorced,their job results are still similar.

```python
plt.figure(figsize=(10,5))
sns.barplot(x="EmpDepartment",y="PerformanceRating",data=data,hue="EmpDepartment")
plt.title("Department VS Performance")
plt.xticks(rotation=45)
plt.tight_layout
```

### INSIGHT
- High performance levels are consistently demonstrated by the Data Science and Development departments.
 

```python
plt.figure(figsize=(10,5))
sns.barplot(x="EmpJobRole",y="PerformanceRating",data=data,hue="EmpJobRole")
plt.title("Job Role VS Performance")
plt.xticks(rotation=45)
plt.tight_layout
```

### INSIGHT
- All individuals regularly exhibit respectable levels of achievement in their performance, but with some differences across different job roles.

```python
plt.figure(figsize=(10,5))
sns.barplot(x="ExperienceYearsAtThisCompany",y="PerformanceRating",data=data,hue="ExperienceYearsAtThisCompany")
plt.title("Experince VS Performance")
plt.tight_layout
```

### INSIGHT
- With twenty-three years of expertise, workers perform at an exceptional level. Notably, the 23rd year sees a notable improvement in performance.

```python
plt.figure(figsize=(20,5))
sns.barplot(x="EmpHourlyRate",y="PerformanceRating",data=data,hue="EmpHourlyRate")
plt.title("Hour Rate VS Performance")
plt.xticks(rotation=90)
plt.tight_layout
```

### INSIGHT
- The performance remains consistent across each hour.

```python
plt.figure(figsize=(10,5))
sns.barplot(x="Attrition",y="PerformanceRating",data=data,hue="Attrition")
plt.title("Attrition VS Performance")
plt.tight_layout
```

### INSIGHT
- Attrition has no effect on performance, which stays constant.

```python
plt.figure(figsize=(10,5))
sns.barplot(x="Age",y="PerformanceRating",data=data,hue="Age")
plt.title("Age VS Performance")
plt.tight_layout
```

### INSIGHT
- Younger individuals exhibit higher performance.

```python
plt.figure(figsize=(10,5))
sns.barplot(x="NumCompaniesWorked",y="PerformanceRating",data=data,hue="NumCompaniesWorked")
plt.title("Companies Worked VS Performance")
plt.tight_layout
```

### INSIGHT
- Performance is higher among workers who have experience in four distinct companies.
